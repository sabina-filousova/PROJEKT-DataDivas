### Toto, je původní soubor z registru vozidel (velká velikost 18 GB)

```python
import pandas as pd

chunksize = 100_000  # velikost jednoho kusu

# připravíme prázdný DataFrame
df_final = pd.DataFrame()

# postupně čteme velký soubor po kouskách
for chunk in pd.read_csv(
    "RSV_vypis_vozidel_20250301.csv",
    usecols=[
        "Datum 1  registrace", "Datum 1  registrace v ČR", "Druh vozidla",
        "Tovární značka", "Palivo", "Plně elektrické vozidlo",
        "Hybridní vozidlo", "PČV", "Rok výroby"
    ],
    sep=",",
    on_bad_lines="skip",
    quoting=0,
    encoding="utf-8",
    low_memory=False,
    chunksize=chunksize  # <- tady říkáme "čti po 100 000 řádcích"
):
    # připojíme vybraný kus do finálního DataFrame
    df_final = pd.concat([df_final, chunk], ignore_index=True)

# na konci uložíme celé zpracované CSV
df_final.to_csv("RSV_vybrane_sloupce2.csv", index=False, encoding="utf-8")
```

```python
from itertools import islice

def print_from_line(filepath, start_line, num_lines):
    """
    Print num_lines lines from filepath, beginning at line index start_line (0-based).
    """
    with open(filepath, 'r', encoding='utf-8') as f:
        # skip up to start_line, then take num_lines
        for line in islice(f, start_line, start_line + num_lines):
            print(line, end='')

print_from_line("RSV_vypis_vozidel_20250301.csv", 200258, 30)
```
